#14 Grouping a DataFrame

```python
df = pd.DataFrame([[1, 2,  "A"], 
                    [5, 8,  "B"], 
                    [3, 10, "B"]], 
                    columns = ["col1", "col2", "col3"])

df.groupby("col3").agg({"col1":"sum", "col2":"max"})
```

#18–21 Filtering a DataFrame

### Method 1: Boolean Filtering

```python
df = pd.DataFrame([[1, 2,  "A"], 
                    [5, 8,  "B"], 
                    [3, 10, "B"]], 
                    columns = ["col1", "col2", "col3"])

print(df[df["col2"] > 5])
```

```python
df = pd.DataFrame([[1, 2,  "A"], 
                   [5, 8,  "B"], 
                   [3, 10, "C"]], 
                  columns = ["col1", "col2", "col3"])

filter_list = ["A", "C"]
print(df[df["col3"].isin(filter_list)])
```

#24 Applying a Function to a DataFrame

```python
def add_cols(row):
    return row.col1 + row.col2

df = pd.DataFrame([[1, 2], 
                   [5, 8], 
                   [3, 9]], 
                  columns = ["col1", "col2"])
                  
df["col3"] = df.apply(add_cols, axis=1)
print(df)
```

```python
def square_col(num):
    return num**2

df = pd.DataFrame([[1, 2], 
                   [5, 8], 
                   [3, 9]], 
                  columns = ["col1", "col2"])
                  
df["col3"] = df.col1.apply(square_col)
print(df)
```

#29 Finding Cross-tabulation

```python
df = pd.DataFrame([["A", "X"], 
                   ["B", "Y"], 
                   ["C", "X"],
                   ["A", "X"]], 
                  columns = ["col1", "col2"])

print(pd.crosstab(df.col1, df.col2))
```

```python
df = pd.DataFrame([["A", "X"], 
                   ["B", "Y"], 
                   ["C", "X"],
                   ["A", "X"]], 
                  columns = ["col1", "col2"])

print(pd.crosstab(df.col1, df.col2))
```
